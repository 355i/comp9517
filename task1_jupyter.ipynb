Import necessary libraries

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2

print(cv2.__version__)
```

Load the input image

```python
input_path = 'Task1.jpg'
input_image = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
```

Define two functions to perform intensity inversion and gamma correction

```python
# Function to perform intensity inversion
def invert_intensity(image):
    return 255 - image

# Function to perform gamma correction
def gamma_correction(image, gamma=1.0):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
    return cv2.LUT(image, table)
```

Execute the above two functions on the input image

```python
inverted_image = invert_intensity(input_image)
gamma_corrected_image = gamma_correction(inverted_image, gamma=0.5)
outputt_image = gamma_corrected_image
```

output and input display

```python
# Display both images side by side for visual comparison
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(input_image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(outputt_image, cmap='gray')
axs[1].set_title('Transformed Image')
axs[1].axis('off')

plt.show()
```
